```python
import pandas as pd
```

```python
path='../../pommermanonevsonecustomallvs1_states_final.csv'
data = pd.read_csv(path,  sep='|', index_col=None).sample(frac=0.5)
```

```python
data
```

```python
small_df = data[['Advisor0_correct', 'Advisor1_correct', 'Advisor2_correct']]
len(small_df[small_df['Advisor0_action'] == small_df['Advisor1_action']]) / len(small_df)
```

```python
df_filtered = small_df[(small_df['Advisor0_correct'] == 1) | (small_df['Advisor1_correct'] == 1)]
df_filtered
```

```python
len(df_filtered[df_filtered['Advisor0_correct'] != df_filtered['Advisor1_correct']]) / len(df_filtered)
```

```python
small_df['sum'] = small_df.apply(sum, axis=1)
small_df
```

```python
filtered_df = small_df[small_df['sum'] > 0]
filtered_df
```

```python
filtered_df['sum'].mean()/4
```
